```python
from sklearn.model_selection import learning_curve
def plot_learning_curve(model, X, y, cv=10, scoring='neg_mean_absolute_error', title=None):
    train_sizes, train_scores, test_scores = learning_curve(
        model, X, y, cv=cv, scoring=scoring)

    # Menghitung nilai rata-rata dari train dan test scores
    train_scores_mean = -np.mean(train_scores, axis=1)
    test_scores_mean = -np.mean(test_scores, axis=1)
    plt.style.use('dark_background')
    plt.figure(figsize=(10, 6))
    plt.plot(train_sizes, train_scores_mean, label='Train', color='royalblue')
    plt.plot(train_sizes, test_scores_mean, label='Test', color='white')
    plt.xlabel('Training Examples')
    plt.ylabel('Score')
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    if title:
        plt.title(title+' based on MAE')
    else:
        plt.title('Learning Curve')
    plt.legend()
    plt.show()

# Memplot learning curve
plot_learning_curve(Linpoly, X_train_poly, y_train, cv=5, scoring='neg_mean_absolute_error', title='Learning Curve - Linpoly Model')
```

```python
plot_learning_curve(rf, X_train, y_train, cv=10, scoring='neg_mean_absolute_error', title='Learning Curve - Random Forest Regressor Model')
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Misalkan Anda sudah memiliki X_train, y_train, X_test, dan y_test

degrees = range(1, 11)  # Degree dari 1 hingga 10
mae_scores = []
mse_scores = []
r2_scores = []

for degree in degrees:
    # Membuat objek PolynomialFeatures dengan degree tertentu
    poly = PolynomialFeatures(degree=degree)
    
    # Transformasi fitur X_train dan X_test sesuai degree polinomial
    X_train_poly = poly.fit_transform(X_train)
    X_test_poly = poly.transform(X_test)
    
    # Membuat model regresi linear dan melatihnya dengan data yang telah di-transformasi
    linreg = LinearRegression()
    linreg.fit(X_train_poly, y_train)
    
    # Melakukan prediksi dengan data test
    y_pred_poly = linreg.predict(X_test_poly)
    
    # Menghitung metrik evaluasi
    mae_score = mean_absolute_error(y_test, y_pred_poly)
    mse_score = mean_squared_error(y_test, y_pred_poly)
    r2_score_val = r2_score(y_test, y_pred_poly)
    
    # Menyimpan nilai metrik evaluasi untuk setiap degree
    mae_scores.append(mae_score)
    mse_scores.append(mse_score)
    r2_scores.append(r2_score_val)

# Membuat DataFrame untuk menyimpan hasil evaluasi
evaluation_df = pd.DataFrame({
    'Degree': degrees,
    'MAE': mae_scores,
    'MSE': mse_scores,
    'R-squared': r2_scores
})

print(evaluation_df)
```
